## 학습
* 1) 문제지 정답지 분리 == target 분리시키기
* 2) 8:2 train test 나누기
* 3) model 선정
* 4) model.fit(문제지8,답안지8) 
* 5) 컴답2 = model.predict(문제지2) 
* 6) accuracy_score(답안지2 ,컴답2)

```python
model=DecisionTreeClassifier(random_state=1234)
model.fit(문제지8,정답지8)
컴퓨터답2 = model.predict(문제지2)
score = accuracy_score(정답지2, 컴퓨터답2)
print("1차점수 : accuracy :", score)
```

```python
model=RandomForestClassifier(n_estimators=20,random_state=1234)
model.fit(문제지8,정답지8)
컴퓨터답2 = model.predict(문제지2)
score = accuracy_score (정답지2, 컴퓨터답2)
score
```

```python
kfold = KFold(n_splits = 3, shuffle=True, random_state=555)
tot_score=[]
for train_index, test_index in kfold.split(iris_data):
#    print(train_index)
#    print(test_index)
#    print("---"*20)
    문제지2, 문제지8 = iris_data.iloc[test_index],iris_data.iloc[train_index]
    정답지2, 정답지8 = iris_data_답안지.iloc[test_index],iris_data_답안지.iloc[train_index]
    
    model = RandomForestClassifier()
    model.fit(문제지8, 정답지8)
    컴퓨터답2 = model.predict(문제지2)
    score = accuracy_score(정답지2, 컴퓨터답2)
    print("fold", score)
    tot_score.append(score)
    
print(tot_score)
print(np.array(tot_score).mean())
```
